```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas as pd
df=pd.read_csv('cardata.csv')
df
```

```python
numeric_cols=['Selling_Price','Year','Present_Price','Kms_Driven',] #selecting numerical columns
corr=df[numeric_cols].corr()#correlation of the columns
sns.heatmap(corr,annot=True)
```

```python
sns.heatmap(corr,annot=True,fmt='1g')#to get whole value
```

```python
#adjusting axis
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0)
```

```python
#colourmap
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral')#Spectral colormap is a perceptually uniform diverging colormap.
#It is commonly used to represent data where both high and low extremes, as well as a midpoint, need to be visually distinct.
```

```python
#linewidth,linecolor
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',linewidth='5',linecolor='black')
```

```python
# orientation
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',cbar_kws={'orientation': 'horizontal'})
```

```python
# orientation
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',cbar_kws={'orientation': 'virtical'})
```

```python
#shape
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',square=True)
```

```python
# changing matrix shape
matrix=np.triu(corr) #triu (lower triangle)
matrix1=np.tril(corr) #tril (upper triangle)
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',mask=matrix)
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',mask=matrix1)
```
