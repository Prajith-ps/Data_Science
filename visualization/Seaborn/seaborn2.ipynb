```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
sns.distplot([0,1,2,3,4,5],hist=False)
plt.show()
```

```python
#normal distribution
x=random.normal(loc=1,scale=2,size=(2,3))
x
```

```python
sns.distplot(random.normal(size=1000),hist=False)
plt.show()
```

```python
#difference between poisson and normal
sns.distplot(random.normal(loc=50,scale=7,size=1000),hist=False,label="normal")
sns.distplot(random.poisson(lam=20,size=1000),hist=False,label="poisson")
plt.legend()
```

```python
# difference between binomial and poisson
sns.distplot(random.binomial(n=1000,p=0.05,size=1100),hist=False,label='binomial')
sns.distplot(random.poisson(lam=20,size=1000),hist=False,label='poisson')
plt.legend()
```
